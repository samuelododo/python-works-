## 4 what is the mean stop duration

```python
# here we'd do a bit of mapping and casting 
police.info()
# from here we can see that the stop_duration is an object and we can't get the mean of an object so we'd need to change that
```

```python
# to convert into an int, we'd have to use mapping, but ist let's check for the distinct values in this column 
police['stop_duration'].value_counts()
# there are just 3 dif distint values so we can easily use the .map method
```

```python
# so now lets convert. as lenght before we can use map to do a reassignment 
police['stop_duration']=police['stop_duration'].map({'0-15 Min':7.5, '16-30 Min':23 , '30+ Min':45,'2':1})
# i used the mean of the mins to give the figure we'd convert it to 
```

```python
police.head(10)
```

```python
police['stop_duration'].mean()
```
